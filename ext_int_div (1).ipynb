```python
import shutil
import os
from ultralytics import YOLO

pretrained_yolo_model = YOLO("exterior/models/yolov8m.pt")
accepted_classes = ["car", "truck", "pickup", "bus"]
data_dir = "ROPRocess"
ext_dir = os.path.join(data_dir, "exterior")
int_dir = os.path.join(data_dir, "interior")
images_list = os.listdir(data_dir)


for i in range(len(images_list)):
    flag = "int"
    image_name = images_list[i]
    if image_name[0]=="." or image_name=="exterior" or image_name=="interior":
        continue
    image_path = os.path.join(data_dir, image_name)
    result = pretrained_yolo_model.predict(image_path, verbose=False)[0]
    if len(result)>0:
        for box in result.boxes:
            class_id = result.names[box.cls[0].item()]
            if class_id in accepted_classes:
                shutil.move(os.path.join(image_path), os.path.join(ext_dir, image_name))
                flag = "ext"
                break
    if flag!="ext":
        shutil.move(os.path.join(image_path), os.path.join(int_dir, image_name))
    print(i, image_name, flag)
```
