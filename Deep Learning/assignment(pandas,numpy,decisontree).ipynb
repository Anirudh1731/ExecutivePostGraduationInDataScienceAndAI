```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Activation Functions

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Step 2: Sample Data
x1 = 3
x2 = 2
y_actual = 17

# Step 3: Feed Forward (initializing random weights)
np.random.seed(42)  # For reproducibility
w1 = np.random.uniform(1, 10)  # Random weight for x1
w2 = np.random.uniform(1, 10)  # Random weight for x2

learning_rate = 0.01
epochs = 20

# Arrays to store error and predicted values for visualization
error_history = []
ypred_history = []

# Step 4: Training loop for 20 epochs
for epoch in range(epochs):
    # Feed Forward: Predict the output using current weights
    y_pred = (x1 * w1) + (x2 * w2)  # y_pred = x1 * w1 + x2 * w2
    
    # Error Calculation: Compute squared error
    error = (y_actual - y_pred) ** 2  # Squared error
    error_history.append(error)
    ypred_history.append(y_pred)
    
    # Step 5: Back Propagation
    # Compute gradients (derivative of error w.r.t. weights)
    grad_w1 = -2 * x1 * (y_actual - y_pred)
    grad_w2 = -2 * x2 * (y_actual - y_pred)
    
    # Update weights using the gradients and learning rate
    w1 -= learning_rate * grad_w1


    w2 -= learning_rate * grad_w2


print(y_pred)
# Step 6: Visualization

# Plot the error across epochs
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(epochs), error_history, label='Squared Error')
plt.title('Error Progression')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.grid(True)

# Plot the predicted output y_pred across epochs
plt.subplot(1, 2, 2)
plt.plot(range(epochs), ypred_history, label='Predicted Output', color='orange')
plt.axhline(y_actual, color='red', linestyle='--', label='Actual Value (y_actual = 17)')
plt.title('Predicted Output Progression')
plt.xlabel('Epochs')
plt.ylabel('y_pred')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
